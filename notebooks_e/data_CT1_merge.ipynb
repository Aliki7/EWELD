```python
# Reset index if 'Time' is already the index
rolling_Ct1_ec_df = rolling_Ct1_ec_df.reset_index()

# Convert 'Time' to datetime and set as index
rolling_Ct1_ec_df['Time'] = pd.to_datetime(rolling_Ct1_ec_df['Time'])
rolling_Ct1_ec_df = rolling_Ct1_ec_df.set_index('Time')

# Calculate rolling max/min over the last 4 periods (1 hour)
rolling_Ct1_ec_df['rolling_hourly_max'] = rolling_Ct1_ec_df['sum energy [kWh]'].rolling(window=4, min_periods=1).max()
rolling_Ct1_ec_df['rolling_hourly_min'] = rolling_Ct1_ec_df['sum energy [kWh]'].rolling(window=4, min_periods=1).min()
```

```python
# shift assignment function
from datetime import datetime, time
def assign_shift(dt):
    t = dt.time()
    if time(6, 0) <= t < time(14, 0):
        return 'First Shift'
    elif time(14, 0) <= t < time(22, 0):
        return 'Second Shift'
    else:
        # Covers 22:00-24:00 and 00:00-06:00
        return 'Third Shift'
```

```python
# copy to new df
Ct1_shift_ec_df = rolling_Ct1_ec_clean_df.copy()
```

```python
# Apply function "assign_shift" to create shift column
Ct1_shift_ec_df ['shift'] = Ct1_shift_ec_df ['DateTime'].apply(assign_shift)
```
